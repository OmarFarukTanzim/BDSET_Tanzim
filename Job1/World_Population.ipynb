# 10. Compute the mean , median, variance and standard deviation if population for each continent

```python
population_columns = [col for col in df.columns if 'Population' in col and 'World Population Percentage' not in col]

for col in population_columns:
    print(f"\nStatistics for {col}:")
    stats_yearly = df.groupby("Continent")[col].agg(
        mean="mean",
        median="median",
        variance="var",
        std_dev="std"
    )
    display(stats_yearly)
```

# 11. Calculate the cumulative population growth over the years for each country

```python
earliest_year_column = population_columns[-1]
earliest_year = earliest_year_column.split(' ')[0]


for col in population_columns:
    if col != earliest_year_column:
        current_year = col.split(' ')[0]
        df[f'Growth_from_{earliest_year}_to_{current_year}'] = df[col] - df[earliest_year_column]

display_columns = ['Country/Territory'] + [f'Growth_from_{earliest_year}_to_{col.split(' ')[0]}' for col in population_columns if col != earliest_year_column]
display(df[display_columns].head())
```

# 12. Chage the column the names


```python
new_columns = {}

for col in df.columns:
    if "Population" in col:
        new_columns[col] = col.split()[0]

df.rename(columns=new_columns, inplace=True)
```

# 14. Visualize continent wise population in 1970 and 2022 using histograms

```python
plt.figure(figsize=(10,5))
sns.histplot(
    data=df,
    x="1970",
    hue="Continent",
    multiple="stack",
    bins=20
)
plt.title("Continent-wise Population Distribution in 1970")
plt.savefig('Continent-wise Population Distribution in 1970_width.png')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(
    data=df,
    x="2022",
    hue="Continent",
    multiple="stack",
    bins=20
)
plt.title("Continent-wise Population Distribution in 2022")
plt.savefig('Continent-wise Population Distribution in 2022_width.png')
plt.show()
```

# 15. Show the timeline fo continent wise population

```python
timeline = df.groupby("Continent")[["1970", "1990", "2000", "2022"]].sum()

timeline.T.plot(figsize=(10,6))
plt.title("Continent-wise Population Growth Timeline")
plt.xlabel("Year")
plt.ylabel("Population")
plt.savefig('Continent-wise Population Growth Timeline.png')
plt.show()
```

# 16. show the correlation between numerical columns

```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])

correlation_matrix = numerical_df.corr()

display(correlation_matrix)

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.savefig('Correlation Matrix of Numerical Columns.png')
plt.show()
```
